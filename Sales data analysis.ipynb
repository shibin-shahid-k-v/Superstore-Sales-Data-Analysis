```python
print(sales.dtypes)
```

```python
sales['Order Date']=sales['Order Date'].astype('datetime64[ns]')
sales['Ship Date']=sales['Ship Date'].astype('datetime64[ns]')
print(sales.dtypes)
```

```python
# Remove duplicate values in the table
sales.drop_duplicates()
```

```python
# Generating descriptive statistics of the data
sales.describe()
```

```python
#Find the unique values in the column of Product Name
sales['Product Name'].nunique()
```

```python
# Identifying the most popular products based on the Product Name and Quantity
ProductName_amount = sales.groupby('Product Name').sum()['Quantity'].sort_values(ascending=False)
ProductName_amount.head(10)
```

```python
#Visualizing the sales on different regions
# Identifying the most popular products based on the Product Name and Sales
ProductName_amount = sales.groupby('Product Name').sum()['Sales'].sort_values(ascending=False)
ProductName_amount.head(10)
```

```python
# Identifying the most popular products based on the Product Name and Profit
ProductName_amount = sales.groupby('Product Name').sum()['Profit'].sort_values(ascending=False)
ProductName_amount.head(10)
```

```python
# Next task is to find how sales revenue vary over time
# For this, table rows has to be categorized based on the year of ordering
# Adding year,month column to the existing table
sales['year'] = pd.DatetimeIndex(sales['Order Date']).year
sales['month'] = pd.DatetimeIndex(sales['Order Date']).month
sales.head(10)
```

```python
#seperate table for Order Date, Sales, Profit, Year, Month, Market
sales_1=sales[['Order Date','Sales' , 'Profit' , 'year' , 'month','Market']]
sales_1
```

```python
sales_market=sales_1.groupby(['Market']).agg({'Sales':'sum'}).reset_index()
sales_market
```

```python
#Visualizing the sales on different regions
sns.barplot(x="Market",y='Sales',data=sales_market)
plt.title('Sales in each market')
```

```python
total_profit_in_years=sales_1.groupby('year')[['Profit']].agg({'Profit':'sum'}).reset_index()
total_profit_in_years
```

```python
sns.lineplot(x="year",y="Profit",data=total_profit_in_years)
plt.title("Toatal Profit in each year")
```

```python
sales_market_year=sales_1.groupby(['year','Market']).agg({"Sales":'sum'}).reset_index()
sales_market_year
```

```python
#Visualizing the sales on different regions year-wise
sns.barplot(x='year',y='Sales',hue='Market',data=sales_market_year)
plt.title('Year-wise Sales')
```

```python
#Calculation of Customer Lifetime Value(CLV) for a period 4 years
# CLV= average purchase value* average purchase frequency* average customer lifespan
# No. of unique customers
sales['Customer ID'].nunique()
```

```python
# Average Purchase Value (APV) = Total Revenue/No. of purchases
APV=sales['Sales'].sum()/sales['Order ID'].count()
print(APV)
```

```python
#Average Purchase Frequency (APF)= number of purchases/number of unique customers
APF=sales['Order ID'].count()/1590
print(APF)
```

```python
print(sales.dtypes)
```

```python
# Average Customer Lifespan (ACL)=  total number of years in the 4-year period / total number of customers. OR
# Average Customer Lifespan (ACL)= 1/ Churn rate
# churn rate= (Customers at the beginning of the period â€“ customers at the end of the period)/customers at the beginning of the period
# Finding unique number of customers at each year
unique_customers_in_year = sales.groupby('year')['Customer ID'].nunique()
unique_customers_in_year
```

```python
unique_customers_in_2011=unique_customers_in_year[2011]
unique_customers_in_2011
```

```python
unique_customers_in_2014=unique_customers_in_year[2014]
unique_customers_in_2014
```

```python
churn_rate=(unique_customers_in_2014-unique_customers_in_2011)/unique_customers_in_2011
churn_rate
```

```python
ACL=1/churn_rate
print(ACL)
```

```python
CLV=ACL*APF*APV
print(CLV)
```
